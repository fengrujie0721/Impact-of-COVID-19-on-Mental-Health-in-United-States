```python
# Collect a list of all the unique values in "date"
anx_data["Time Period Start Date"].unique()
```

```python
# Looking only at 04/23/2020
anx_042320_df = anx_data.loc[anx_data["Time Period Start Date"] == "4/23/2020 0:00", :]
anx_042320_df
```

```python
#Calculate average anxiety level as of 04/23/20
avg_anx_042320 = (anx_042320_df["Value"]).mean()
avg_anx_042320_df = pd.DataFrame({"Average American anxiety level as of 04/23/20": [avg_anx_042320]})
avg_anx_042320_df
```
