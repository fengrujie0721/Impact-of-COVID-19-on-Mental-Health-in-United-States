```python
import pandas as pd
import matplotlib.pyplot as plt
# File to Load
file_to_load = "Resources/us_states.csv"

# Read CSV File and store into Pandas data frame
covid_data = pd.read_csv(file_to_load)
covid_data.head()
```

```python
del covid_data['fips']
covid_data.head()
```

```python
total_states = len(covid_data["state"].value_counts())
states_count = pd.DataFrame({"Total States": [total_states]})
states_count
```

```python
# Collect a list of all the unique values in "date"
covid_data["date"].unique()
```

```python
# Looking only at 03/15/2020 which is early-point
covid_031520_df = covid_data.loc[covid_data["date"] =="3/15/2020", :]
covid_031520_df.head()
```

```python
# Sort the DataFrame by "deaths" in ascending order
covid_031520_df = covid_031520_df.sort_values("deaths")
covid_031520_df
```

```python
# Calculate total deaths as of 03/15/2020
total_deaths_031520 = (covid_031520_df["deaths"]).sum()
total_deaths_031520_df = pd.DataFrame({"Total deaths from COVID-19 as of 03/15/2020": [total_deaths_031520]})
total_deaths_031520_df
```

```python
# Looking only at 10/01/2020 which is mid-point
covid_100120_df = covid_data.loc[covid_data["date"] =="10/1/2020", :]
covid_100120_df.head()
```

```python
# Sort the DataFrame by "deaths" in ascending order
covid_100120_df = covid_100120_df.sort_values("deaths")
covid_100120_df
```

```python
# Calculate total deaths as of 10/01/2020
total_deaths_100120 = (covid_100120_df["deaths"]).sum()
total_deaths_100120_df = pd.DataFrame({"Total deaths from COVID-19 as of 10/01/2020": [total_deaths_100120]})
total_deaths_100120_df
```

```python
# Looking only at 4/25/2021 which is peak
covid_042521_df = covid_data.loc[covid_data["date"] =="4/25/2021", :]
covid_042521_df.head()
```

```python
# Sort the DataFrame by "deaths" in ascending order
covid_042521_df = covid_042521_df.sort_values("deaths")
covid_042521_df
```

```python
# Calculate total deaths as of 04/25/2021
total_deaths_042521 = (covid_042521_df["deaths"]).sum()
total_deaths_042521_df = pd.DataFrame({"Total deaths from COVID-19 as of 04/25/2021": [total_deaths_042521]})
total_deaths_042521_df
```

```python
chart_042521 = pd.DataFrame(covid_042521_df.groupby(["state"]).count()).reset_index()
chart_042521_df = chart_042521[["state", "deaths"]]
chart_042521_df = chart_042521_df.set_index("state")

chart_042521_df.plot(kind="bar", figsize=(20,10))
plt.title("Total deaths from COVID-19 as of 04/25/2021")
plt.show()
```
